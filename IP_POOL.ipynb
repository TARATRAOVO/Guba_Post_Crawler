```python
import time
import requests
import pandas as pd
```

```python
class IP_POOL:
    def __init__(self, ip_pool_location):
        self.ip_pool_location = ip_pool_location
        self.all_ip_keys = []
        self.ip_pool = pd.DataFrame(columns=('IP', 'TIME', 'COUNT'))
        while True:
            try:
                self.ip_pool.to_csv(self.ip_pool_location, index=False)
                break
            except:
                continue
        ip_pool_location = []
        pass

    def add_ip_keys(self, ip_keys):
        for i in ip_keys:
            self.all_ip_keys.append(i)

    def print_ip_keys(self):
        print(self.all_ip_keys)

    def get_ip(self, key):
        dynamic_ip = requests.get(key)
        return 'http://' + dynamic_ip.text

    def refresh_ip_pool(self):
        TIME = time.time()
        ip_pool = self.ip_pool
        self.ip_pool = ip_pool.drop(ip_pool[ip_pool.TIME - TIME < -60].index)
    


    def update_ip_pool(self):
        while True:
            for key in self.all_ip_keys:

                TIME = time.time()
                try: 
                    ip = self.get_ip(key)
                except:
                    time.sleep(10/len(self.all_ip_keys))
                    continue
                self.new_ip = pd.DataFrame()
                self.new_ip['IP'] = [ip]
                self.new_ip['TIME'] = [TIME]
                self.new_ip['COUNT'] = [0]
                while True:
                    try:
                        self.ip_pool = pd.read_csv(self.ip_pool_location)
                        break
                    except:
                        continue
                    
                self.ip_pool = pd.concat([self.ip_pool,self.new_ip], ignore_index=True)
                self.refresh_ip_pool()
                while True:
                    try:
                        self.ip_pool.to_csv(self.ip_pool_location, index=False)
                        break
                    except:
                        continue
                time.sleep(10/len(self.all_ip_keys))




keys = ['https://api.xiaoxiangdaili.com/ip/get?appKey=836222203564806144&appSecret=c80Pi6Es&cnt=&wt=text']
ip_pool = IP_POOL('ip_pool.csv')
ip_pool.add_ip_keys(keys)
ip_pool.update_ip_pool()
```
